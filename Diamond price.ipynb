```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('D:\Diamondprice.csv')
df.head()
```

```python
df.info()
```

```python
plt.title('Diamond Price with Increasing Carat Weight')
sns.distplot(df['carat'])
plt.show()
```

```python
plt.scatter(df['carat'], df['price'], color = 'lightcoral')
plt.title('carat Vs price')
plt.xlabel('carat')
plt.ylabel('price')
plt.box(False)
plt.show()
```

```python
X = df[['carat']]  
y = df['price'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train)
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('carat Vs price')
plt.xlabel('carat')
plt.ylabel('price')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'carat/price', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('carat Vs price')
plt.xlabel('carat')
plt.ylabel('pprice')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'carat/price', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
